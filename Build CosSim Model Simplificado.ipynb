```python
import glob
import gensim.models.word2vec as w2v
import multiprocessing
```

```python
controlDocsList = sorted(glob.glob("C:/Users/danie/Dropbox/Proyecto Psi - Mauricio/Texto/Train Model/*.txt"))

def createNamesList(docsList):
    namesList = []
    for doc in docsList:
        namesList.append(doc[-11:-4])
    return(namesList)

controlNames = createNamesList(controlDocsList)
```

```python
len(controlDocsList)
```

```python
def createDicts(namesList, etiqueta):
    dictionary = {}
    for name in namesList:
        dictionary[name + etiqueta] = {}
    return(dictionary)

controlLemSinSWTxtDicts = createDicts(controlNames, "_texto")
```

```python
def fillTxtDictionaries(fileNamesList, namesList, txtDicts):
    for i in range(0, len(fileNamesList)):
        indice_E = 1
        indice_I = 1
        leyendo = True
        with open(fileNamesList[i]) as txt:
            for line in txt:
                if leyendo:
                    txtDicts[namesList[i]+"_texto"][line[0] + str(indice_E)] = line[3:].split()
                    indice_E += 1
                else:
                    txtDicts[namesList[i]+"_texto"][line[0] + str(indice_I)] = line[3:].split()
                    indice_I += 1
                leyendo = not leyendo
                
fillTxtDictionaries(controlDocsList, controlNames, controlLemSinSWTxtDicts)
```

```python
def buildCorpus(txtDicts):
    corpus = []
    for keys, values in txtDicts.iteritems():
        for key, value in values.iteritems():
            corpus.append(value)
    return(corpus)

corpus = buildCorpus(controlLemSinSWTxtDicts)
```

```python
num_features = 300
min_word_count = 3
num_workers = multiprocessing.cpu_count()
context_size = 7
downsampling = 1e-3
seed = 1

modelo = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling)

modelo.build_vocab(corpus)
modelo.train(corpus, total_examples = modelo.corpus_count, epochs = 5)
```

```python
def nearest_similarity_cosmul(start1, end1, end2):
    similarities = modelo.most_similar_cosmul(
        positive=[end2, start1],
        negative=[end1]
    )
    start2 = similarities[0][0]
    print("{start1} se relaciona con {end1}, así como {start2} se relaciona con {end2}".format(**locals()))
    return start2

nearest_similarity_cosmul("polola", "pololo", "papá")
nearest_similarity_cosmul("pavo", "navidad", "dieciocho")
nearest_similarity_cosmul("auto", "toyota", "ekono")
nearest_similarity_cosmul("hermano", "hermana", "mujer")
```
