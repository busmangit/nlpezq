```python
def createDicts(namesList, etiqueta):
    dictionary = {}
    for name in namesList:
        dictionary[name + etiqueta] = {}
    return(dictionary)

controlLemSinExpTxtDicts = createDicts(controlNames, "_texto")
primEpLemSinExpTxtDicts = createDicts(primEpNames, "_texto")
cronicoLemSinExpTxtDicts = createDicts(cronicoNames, "_texto")

controlSinExpTxtDicts = createDicts(controlNames, "_texto")
primEpSinExpTxtDicts = createDicts(primEpNames, "_texto")
cronicoSinExpTxtDicts = createDicts(cronicoNames, "_texto")

controlSinAllSWTxtDicts = createDicts(controlNames, "_texto")
primEpSinAllSWTxtDicts = createDicts(primEpNames, "_texto")
cronicoSinAllSWTxtDicts = createDicts(cronicoNames, "_texto")
```

```python
def fillTxtDictionaries(fileNamesList, namesList, txtDicts):
    for i in range(0, len(fileNamesList)):
        indice_E = 1
        indice_I = 1
        leyendo = True
        with open(fileNamesList[i], encoding="utf-8") as txt:
            for line in txt:
                if leyendo:
                    txtDicts[namesList[i]+"_texto"][line[0] + str(indice_E)] = line[3:]
                    indice_E += 1
                else:
                    txtDicts[namesList[i]+"_texto"][line[0] + str(indice_I)] = line[3:]
                    indice_I += 1
                leyendo = not leyendo
                
fillTxtDictionaries(controlLemSinExpDocsList, controlNames, controlLemSinExpTxtDicts)
fillTxtDictionaries(primEpLemSinExpDocsList, primEpNames, primEpLemSinExpTxtDicts)
fillTxtDictionaries(cronicoLemSinExpDocsList, cronicoNames, cronicoLemSinExpTxtDicts)

fillTxtDictionaries(controlSinExpDocsList, controlNames, controlSinExpTxtDicts)
fillTxtDictionaries(primEpSinExpDocsList, primEpNames, primEpSinExpTxtDicts)
fillTxtDictionaries(cronicoSinExpDocsList, cronicoNames, cronicoSinExpTxtDicts)

fillTxtDictionaries(controlSinAllSWDocsList, controlNames, controlSinAllSWTxtDicts)
fillTxtDictionaries(primEpSinAllSWDocsList, primEpNames, primEpSinAllSWTxtDicts)
fillTxtDictionaries(cronicoSinAllSWDocsList, cronicoNames, cronicoSinAllSWTxtDicts)
```

```python
controlDataDicts = createDicts(controlNames, "_data")
primEpDataDicts = createDicts(primEpNames, "_data")
cronicoDataDicts = createDicts(cronicoNames, "_data")

def fillDataDicts(txtDicts, namesList, dataDicts):
    palabrasTotalesPorOracion = []
    for name in namesList:
        palabrasTotales = 0
        palabrasDiferentes = []
        listPalabrasTotalesPorOracion = []
        listPalabrasDiferentesPorOracion = []
        for i in range(0, len(txtDicts[name + "_texto"])//2):
            sentence = txtDicts[name + "_texto"]["I" + str(i+1)].split()
            listPalabrasTotalesPorOracion.append(len(sentence))
            palabrasDiferentesPorOracion = []
            for word in sentence:
                palabrasTotales += 1
                if word not in palabrasDiferentesPorOracion:
                    palabrasDiferentesPorOracion.append(word)
                if word not in palabrasDiferentes:
                    palabrasDiferentes.append(word)
            listPalabrasDiferentesPorOracion.append(len(palabrasDiferentesPorOracion))
        dataDicts[name + "_data"]["PalabrasTotalesPorOracion"] = listPalabrasTotalesPorOracion
        dataDicts[name + "_data"]["PalabrasDiferentesPorOracion"] = listPalabrasDiferentesPorOracion
        dataDicts[name + "_data"]["PalabrasDiferentes"] = len(palabrasDiferentes)
        dataDicts[name + "_data"]["PalabrasTotales"] = palabrasTotales
        dataDicts[name + "_data"]["ParesPregResp"] = len(txtDicts[name + "_texto"])/2
        
fillDataDicts(controlLemSinExpTxtDicts, controlNames, controlDataDicts)
fillDataDicts(primEpLemSinExpTxtDicts, primEpNames, primEpDataDicts)
fillDataDicts(cronicoLemSinExpTxtDicts, cronicoNames, cronicoDataDicts)
```

```python
PronPersonales = ["yo","me","mí","nos","nosotras","nosotros","conmigo","te","ti","tú","os","usted","vos","vosotras","vosotros","contigo","él","ella","ellas","ello","ellos","la","las","lo","los","le","les","se","sí","consigo"]
PronDemostrativos = ["aquéllas","aquélla","aquéllos","aquél","aquellas","aquella","aquellos","aquel","aquello","ésas","ésa","esas","esa","esos","ese","ésos","ése","eso","esotra","esotro","esta","éstas","ésta","estas","estos","este","éstos","éste","esto","estotra","estotro"]
PronPosesivos = ["mía","mías","mío","míos","nuestra","nuestras","nuestro","nuestros","suya","suyas","suyo","suyos","tuya","tuyas","tuyo","tuyos","vuestra","vuestras","vuestro","vuestros"]
PronIndefinidos = ["algo","alguien","alguna","algunas","alguno","algunos","cualesquiera","cualquiera","demás","misma","mismas","mismo","mismos","mucha","muchas","mucho","muchos","nada","nadie","ninguna","ningunas","ninguno","ningunos","otra","otras","otro","otros","poca","pocas","poco","pocos","quienquier","quienesquiera","quienquiera","tanta","tantas","tanto","tantos","toda","todo","todas","todos","última","últimas","último","últimos","una","unas","uno","unos","varias","varios"]
PronInterrogativos = ["adónde","cómo","cuál","cuáles","cuándo","cuánta","cuántas","cuánto","cuántos","dónde","qué","quién","quiénes"]
PronRelativos = ["como","donde","cuando","cual","cuales","cuanta","cuantas","cuantos","cuanto","cuya","cuyas","cuyo","cuyos","que","quien","quienes"]
```

```python
def fillSintaxData(txtDicts, namesList, dataDicts):
    for name in namesList:
    
        PronPersonalesCounter = 0
        PronDemostrativosCounter = 0
        PronPosesivosCounter = 0
        PronIndefinidosCounter = 0
        PronInterrogativosCounter = 0
        PronRelativosCounter = 0
    
        DetDemostrativosCounter = 0
        DetPosesivosCounter = 0
        DetInterrogativosCounter = 0
        DetIndefinidosCounter = 0
    
        for i in range(0, len(txtDicts[name + "_texto"])//2):
            sentence = txtDicts[name + "_texto"]["I" + str(i+1)].split()
            for word in sentence:
            
                if word in PronPersonales:
                    PronPersonalesCounter += 1
                if word in PronDemostrativos:
                    PronDemostrativosCounter += 1
                if word in PronPosesivos:
                    PronPosesivosCounter += 1
                if word in PronIndefinidos:
                    PronIndefinidosCounter += 1
                if word in PronInterrogativos:
                    PronInterrogativosCounter += 1
                if word in PronRelativos:
                    PronRelativosCounter += 1
                
                if word in DetDemostrativos:
                    DetDemostrativosCounter += 1
                if word in DetPosesivos:
                    DetPosesivosCounter += 1
                if word in DetInterrogativos:
                    DetInterrogativosCounter += 1
                if word in DetIndefinidos:
                    DetIndefinidosCounter += 1
        
        dataDicts[name + "_data"]["numPronPersonales"] = PronPersonalesCounter
        dataDicts[name + "_data"]["numPronDemostrativos"] = PronDemostrativosCounter
        dataDicts[name + "_data"]["numPronPosesivos"] = PronPosesivosCounter
        dataDicts[name + "_data"]["numPronIndefinidos"] = PronIndefinidosCounter
        dataDicts[name + "_data"]["numPronInterrogativos"] = PronInterrogativosCounter
        dataDicts[name + "_data"]["numPronRelativos"] = PronRelativosCounter
        dataDicts[name + "_data"]["numPronTotal"] = PronPersonalesCounter+PronDemostrativosCounter+PronPosesivosCounter+PronIndefinidosCounter+PronInterrogativosCounter+PronRelativosCounter
    
        dataDicts[name + "_data"]["numDetDemostrativos"] = DetDemostrativosCounter
        dataDicts[name + "_data"]["numDetPosesivos"] = DetPosesivosCounter
        dataDicts[name + "_data"]["numDetInterrogativos"] = DetInterrogativosCounter
        dataDicts[name + "_data"]["numDetIndefinidos"] = DetIndefinidosCounter
        dataDicts[name + "_data"]["numDetTotal"] = DetDemostrativosCounter+DetPosesivosCounter+DetInterrogativosCounter+DetIndefinidosCounter

fillSintaxData(controlSinExpTxtDicts, controlNames, controlDataDicts)
fillSintaxData(primEpSinExpTxtDicts, primEpNames, primEpDataDicts)
fillSintaxData(cronicoSinExpTxtDicts, cronicoNames, cronicoDataDicts)
```

```python
def fillContentData2(txtDicts, namesList, dataDicts):
    for name in namesList:
        palabrasDiferentes = []
        total = 0
        for i in range(0, len(txtDicts[name + "_texto"])//2):
            sentence = txtDicts[name + "_texto"]["I" + str(i+1)].split()
            for word in sentence:
                total += 1
                if word not in palabrasDiferentes:
                    palabrasDiferentes.append(word)
                if total % 250 == 0:
                    dataDicts[name + "_data"]["PalabrasDiferentespor" + str(total) + "Palabras"] = len(palabrasDiferentes)
            if total >= 1000:
                break

fillContentData2(controlLemSinExpTxtDicts, controlNames, controlDataDicts)
fillContentData2(primEpLemSinExpTxtDicts, primEpNames, primEpDataDicts)
fillContentData2(cronicoLemSinExpTxtDicts, cronicoNames, cronicoDataDicts)
```

```python
def fillLargoDePalabras(txtDicts, namesList, dataDicts, text):
    for name in namesList:
        largoDePalabras = []
        for i in range(0, len(txtDicts[name + "_texto"])//2):
            sentence = txtDicts[name + "_texto"]["I" + str(i+1)].split()
            for word in sentence:
                k = 0
                for letter in word:
                    k+=1
                largoDePalabras.append(k)
        dataDicts[name + "_data"]["LargoDePalabras" + text] = largoDePalabras

fillLargoDePalabras(controlSinExpTxtDicts, controlNames, controlDataDicts, "SinExp")
fillLargoDePalabras(primEpSinExpTxtDicts, primEpNames, primEpDataDicts, "SinExp")
fillLargoDePalabras(cronicoSinExpTxtDicts, cronicoNames, cronicoDataDicts, "SinExp")

fillLargoDePalabras(controlSinAllSWTxtDicts, controlNames, controlDataDicts, "SinAllSW")
fillLargoDePalabras(primEpSinAllSWTxtDicts, primEpNames, primEpDataDicts, "SinAllSW")
fillLargoDePalabras(cronicoSinAllSWTxtDicts, cronicoNames, cronicoDataDicts, "SinAllSW")
```

```python
controlFinalDicts = createDicts(controlNames, "")
primEpFinalDicts = createDicts(primEpNames, "")
cronicoFinalDicts = createDicts(cronicoNames, "")

def fillFinalDicts(namesList, dataDicts, finalDicts, Clase):
    for name in namesList:
        print(name)
        finalDicts[name]["Clase"] = Clase
        finalDicts[name]["Promedio_Palabras_Totales_Por_Respuesta"] = np.mean(dataDicts[name + "_data"]["PalabrasTotalesPorOracion"])
        finalDicts[name]["Promedio_Palabras_Diferentes_Por_Respuesta"] = np.mean(dataDicts[name + "_data"]["PalabrasDiferentesPorOracion"])
        finalDicts[name]["Pronombres_Personales_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronPersonales"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Pronombres_Demostrativos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronDemostrativos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Pronombres_Posesivos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronPosesivos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Pronombres_Indefinidos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronIndefinidos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Pronombres_Interrogativos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronInterrogativos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Pronombres_Relativos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronRelativos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Pronombres_Totales_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numPronTotal"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Determinantes_Demostrativos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numDetDemostrativos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Determinantes_Posesivos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numDetPosesivos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Determinantes_Interrogativos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numDetInterrogativos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Determinantes_Indefinidos_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numDetIndefinidos"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Determinantes_Totales_Por_Palabras_Totales"] = dataDicts[name + "_data"]["numDetTotal"]/float(dataDicts[name + "_data"]["PalabrasTotales"])
        finalDicts[name]["Largo_Promedio_de_Palabras"] = np.mean(dataDicts[name + "_data"]["LargoDePalabrasSinExp"])
        finalDicts[name]["Largo_Promedio_de_Palabras_Sin_SW"] = np.mean(dataDicts[name + "_data"]["LargoDePalabrasSinAllSW"])
        finalDicts[name]["Pares_Preg-Rpta_Por_Tiempo"] = dataDicts[name + "_data"]["ParesPregResp"]/dataDicts[name + "_data"]["Tiempo"]
        finalDicts[name]["Palabras_Diferentes_Por_Tiempo"] = dataDicts[name + "_data"]["PalabrasDiferentes"]/dataDicts[name + "_data"]["Tiempo"]
        finalDicts[name]["Palabras_Totales_Por_Tiempo"] = dataDicts[name + "_data"]["PalabrasTotales"]/dataDicts[name + "_data"]["Tiempo"]
        try:
            finalDicts[name]["TTR250"] = dataDicts[name + "_data"]["PalabrasDiferentespor250Palabras"]/float(250)
            finalDicts[name]["TTR500"] = dataDicts[name + "_data"]["PalabrasDiferentespor500Palabras"]/float(500)
            finalDicts[name]["TTR750"] = dataDicts[name + "_data"]["PalabrasDiferentespor750Palabras"]/float(750)
            finalDicts[name]["TTR1000"] = dataDicts[name + "_data"]["PalabrasDiferentespor1000Palabras"]/float(1000)
        except (KeyError):
            pass

fillFinalDicts(controlNames, controlDataDicts, controlFinalDicts, "CON")
fillFinalDicts(primEpNames, primEpDataDicts, primEpFinalDicts, "PEE")
fillFinalDicts(cronicoNames, cronicoDataDicts, cronicoFinalDicts, "CRE")
```
